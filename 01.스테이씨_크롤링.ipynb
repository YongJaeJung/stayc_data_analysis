```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver import ActionChains

import pandas as pd


from time import sleep
from tqdm import tqdm

keyword = '스테이씨'
url = 'https://www.youtube.com/results?search_query={}'.format(keyword)
s = Service('./chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get(url)
#ActionChains생성
action = ActionChains(driver)
```

```python
for i in tqdm(range(30)):
    start = i * 10000
    end = (i+1) * 10000
    driver.execute_script(f"window.scrollTo({start},{end})")
    sleep(2)
```

```python
stayc_titles = driver.find_elements(By.ID, "video-title")
```

```python
stayc_uploader= driver.find_elements(By.XPATH, "//a[@class='yt-simple-endpoint style-scope yt-formatted-string']")
```

```python
titles = [x.text for x in stayc_titles]
uploaders = [stayc_uploader[i].text for i in range(len(stayc_uploader)) if i % 2 == 1]
```

```python
driver.execute_script(f"window.scrollTo(700,0)")
```

```python
stayc_list = []

views = []
update_dates = []
    
for i,title in enumerate(stayc_titles):
    try:
        title.send_keys(Keys.CONTROL +"\n")  
        driver.switch_to.window(driver.window_handles[1])
        sleep(2)

        stayc_view = driver.find_element(By.XPATH, "//span[@class='view-count style-scope ytd-video-view-count-renderer']").text
        stayc_updt_date = driver.find_element(By.XPATH, "//div[@id='info-strings']").text
        print(stayc_view,stayc_updt_date)
        views.append(stayc_view)
        update_dates.append(stayc_updt_date)
        
        sleep(1)
        # 현재 사용중인 탭 종료
        driver.close()

        # 메인 탭으로 이동
        driver.switch_to.window(driver.window_handles[0])
        sleep(1)
    except:
        print(i)

        
for title,uploader,view,update_date in zip(titles,uploaders,views,update_dates): 
    stayc_list.append([title,uploader,view,update_date])
```

```python
stayc_list
```

```python
stayc_df = pd.DataFrame(data=stayc_list,columns=['제목','업로더','조회수','게시일'])
```

```python
stayc_df = stayc_df.drop_duplicates(keep='last')
```

```python
stayc_df=  stayc_df.reset_index(drop=True)
```
